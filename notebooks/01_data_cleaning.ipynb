```python
from scipy.stats import ttest_ind

for col in df.columns:
    if col != "Injury_Risk":
        group0 = df[df["Injury_Risk"] == 0][col]
        group1 = df[df["Injury_Risk"] == 1][col]
        
        stat, p = ttest_ind(group0, group1)
        print(f"{col}: p-value = {p}")
```

```python
import numpy as np

def cohens_d(x, y):
    return (x.mean() - y.mean()) / np.sqrt((x.std()**2 + y.std()**2) / 2)

for col in df.columns:
    if col != "Injury_Risk":
        group0 = df[df["Injury_Risk"] == 0][col]
        group1 = df[df["Injury_Risk"] == 1][col]
        
        d = cohens_d(group0, group1)
        print(f"{col}: Cohen's d = {d}")
```

Bivariate Analysis Summary
Independent sample t-tests revealed that Injury_History, Sleep_Hours, Muscle_Asymmetry, and Training_Intensity are significantly associated with Injury_Risk (p < 0.05). Among these, Injury_History and Sleep_Hours show the strongest effect sizes, with Sleep_Hours demonstrating a moderate negative effect (Cohens d  0.52). Other variables such as Age, BMI, Height, and Weight do not show statistically significant differences between risk groups. These findings indicate that prior injuries, training load, muscular imbalance, and recovery are the primary drivers of injury risk in this dataset.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

# escludiamo la target
X = df.drop(columns=["Injury_Risk"])

vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

vif_data.sort_values(by="VIF", ascending=False)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

X = df.drop(columns=["Injury_Risk", "BMI"])

vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) 
                   for i in range(X.shape[1])]

vif_data.sort_values(by="VIF", ascending=False)
```

Multivariate Analysis Summary
Variance Inflation Factor (VIF) analysis revealed substantial multicollinearity among several predictors, particularly Height_cm, Weight_kg, and Sleep_Hours. This is likely due to structural relationships within the dataset (e.g., BMI derivation and training-related dependencies). While multicollinearity may affect linear models, it does not significantly impact tree-based algorithms. Therefore, no immediate feature removal was applied, and the dataset remains suitable for machine learning modeling.
